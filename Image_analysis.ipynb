**<h4> 'L' indicates that the image is in greyscale </h4>**

Other image modes:


*   P: 8-bit color palette
*   RGB: 3 channels with 8-bit each - true color


*   RGBA: 4 channels with 8-bit each - true color with transparency mask
*   CMYk: 4 channels with 8-bit each - color separation

and others

```python
#Checking the size of the image

xray_image.size
```

```python
plt.figure(figsize= (8,8))

plt.imshow(xray_image, cmap = 'gray')

plt.title('Chest X-Ray Image')

plt.axis('off')

plt.show()
```

```python
image_matrix = np.array(xray_image)
image_matrix
```

```python
plt.figure(figsize= (8,6))

plt.hist(image_matrix.ravel(), bins = 256)

plt.title('Pixel intensity distribution of Chest X-Ray Image')

plt.xlabel('Pixel Intensity')

plt.ylabel('Frequency')

plt.grid(True)

plt.show()
```

```python
flattened_matrix = image_matrix.flatten()
flattened_matrix
```

```python
#Make image smaller

resized_img = xray_image.resize((256,256))
resized_img
```

```python
#Normalization of the image intensity values

normalized_matrix = image_matrix/255.0
normalized_matrix
```
